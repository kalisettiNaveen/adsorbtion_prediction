```python
import pandas as pd
import numpy as np
import seaborn as sns
import pickle
```

```python
df = pd.read_csv('absorbance.csv')
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df = df.select_dtypes(include =['int64','float64']) 
df
```

```python
newdf = df.fillna({
    '5ml/min' :0,
    '10ml/min' :0,
    '15ml/min' :0,
    '20ml/min' :0,
})
```

```python
newdf
```

```python
x = newdf[['time']].values
y1 = newdf['5ml/min'].values
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y1_train,y1_test = train_test_split(x,y1,random_state=5)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt1=DecisionTreeRegressor(random_state=0,criterion="absolute_error")
dt1.fit(x_train,y1_train)
pre1 = dt1.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y1_test,pre1)
```

```python
print(dt1.predict([[30]]))
```
